```python
#### divide the into independent and dependent varaibles
```

```python
x=df.drop(["Id","SalePrice"],axis=1)
y=df["SalePrice"]
```

```python
##### split  the  data into trainset and testset by using split method
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.2,random_state=1212121)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape 
```

```python
dftraincat=x_train.select_dtypes(object)
dftrainnum=x_train.select_dtypes(np.number)
```

```python
dftraincat.drop(["MiscFeature", "Fence", "PoolQC", "Alley"], 
                          axis=1, inplace=True)
```

```python
from sklearn.impute import SimpleImputer
sImputer = SimpleImputer(strategy="most_frequent")
```

```python
sImputer.fit(dftraincat)
```

```python
temp=sImputer.transform(dftraincat)
```

```python
dftraincatnalremoved=pd.DataFrame(temp,columns=dftraincat.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dftraincatnalremoved.shape
```

```python
dftraincatencoded=dftraincatnalremoved.apply(le.fit_transform)
```

```python
dftraincatencoded.shape
```

```python
####### train numrical
```

```python
dftrainnum.shape
```

```python
sImputerNum = SimpleImputer(strategy="mean")
```

```python
sImputerNum.fit(dftrainnum)
```

```python
nume=sImputerNum.transform(dftrainnum)
```

```python
dftrainnumnalremoved=pd.DataFrame(nume,columns=dftrainnum.columns)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
dftrainnumscaled=pd.DataFrame(ss.fit_transform(dftrainnumnalremoved))
```

```python
##### concatination of the train set
```

```python
finaltrainset=pd.concat([dftrainnumscaled,dftraincatencoded],axis=1)
```

```python
dftrainnumscaled.shape,dftraincatencoded.shape
```

```python
finaltrainset.isna().sum()
```

```python
###### processing of test test
```

```python
dftestcat=x_test.select_dtypes(object)
dftestnum=x_test.select_dtypes(np.number)
```

```python
dftestcat.shape
```

```python
dftestcat.drop(["MiscFeature", "Fence", "PoolQC", "Alley"], 
                          axis=1, inplace=True)
```

```python
sImputer.fit(dftestcat)
```

```python
teatca=sImputer.transform(dftestcat)
```

```python
testcatanalremoved=pd.DataFrame(teatca,columns=dftestcat.columns)
```

```python
testcatencoded=testcatanalremoved.apply(le.fit_transform)
```

```python
##### test numerical
```

```python
testnum=sImputerNum.transform(dftestnum)
```

```python
dftestnumnalremoved=pd.DataFrame(testnum,columns=dftestnum.columns)
```

```python
testnum=ss.transform(dftestnumnalremoved)
```

```python
dftestcatscaled=pd.DataFrame(testnum,columns=dftestnumnalremoved.columns)
```

```python
##### concatination of the test set
```

```python
finaltest=pd.concat([dftestcatscaled,testcatencoded],axis=1)
```

```python
###### model evaluation
```

```python
from sklearn.linear_model import LinearRegression,SGDRegressor,Ridge,ElasticNet,Lasso
```

```python
lr=LinearRegression()
sdescentreg1=SGDRegressor()
lasso1=Lasso(alpha=0.5)
ridge1=Ridge(alpha=0.5)
enet1=ElasticNet(l1_ratio=0.4)
```

```python
lr.fit(finaltrainset,y_train)
```

```python
sdescentreg1.fit(finaltrainset,y_train)
```

```python
lasso1.fit(finaltrainset,y_train)
```

```python
ridge1.fit(finaltrainset,y_train)
```

```python
enet1.fit(finaltrainset,y_train)
```

```python
finaltrainset.shape,finaltest.shape
```

```python
#### predictions
```

```python
lrpred=lr.predict(finaltest)
```

```python
lasso1pred=lasso1.predict(finaltest)
```

```python
ridge1pred=ridge1.predict(finaltest)
```

```python
sdescentreg1pred=sdescentreg1.predict(finaltest)
```

```python
enet1pred=enet1.predict(finaltest)
```

```python
####  import the metrics
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print(np.sqrt(mean_squared_error(lrpred,y_test)))
print(r2_score(lrpred,y_test))
```

```python
print(np.sqrt(mean_squared_error(lasso1pred,y_test)))
print(r2_score(lasso1pred,y_test))
```

```python
print(np.sqrt(mean_squared_error(ridge1pred,y_test)))
print(r2_score(ridge1pred,y_test))
```

```python
print(np.sqrt(mean_squared_error(sdescentreg1pred,y_test)))
print(r2_score(sdescentreg1pred,y_test))
```

```python
print(np.sqrt(mean_squared_error(enet1pred,y_test)))
print(r2_score(enet1pred,y_test))
```
